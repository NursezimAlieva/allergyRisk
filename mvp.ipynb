```python
print('Hello, here is simple way to know risk of getting som allergy.')
print('Choose allergy: 0-food, 1-pollen, 2-dust')
```

```python
allergy = int(input('Choose a number: '))
if allergy == 0:
    familyH = input('Does your parents have food allergy?(Write Yes or No)')
    if familyH == 'Yes':
        print('Risk of food allergy is high.')
    else:
        print('Risk of food allergy is low.')
elif allergy == 1:
    asthma = input('Do you have asthma?(Write Yes or No)')
    if asthma == 'Yes':
        print('Risk of pollen allergy is high.')
    else: 
        print('Risk of pollen allergy is low.')
else:
    print('Wait for another allergy.')
```
