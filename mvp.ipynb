```python
print('Hello, here is simple way to know risk of getting som allergy.')
print('Choose allergy: 0-food, 1-pollen, 2-dust, 3-pet')
```

```python
allergy = int(input('Choose a number: '))
if allergy == 0:
    familyH = input('Does your parents have food allergy?(Write Yes or No)')
    if familyH == 'Yes':
        print('Risk of food allergy is high.=(')
    else:
        print('Risk of food allergy is low.=)')
elif allergy == 1:
    asthma = input('Do you have asthma?(Write Yes or No)')
    if asthma == 'Yes':
        print('Risk of pollen allergy is high.=(')
    else: 
        print('Risk of pollen allergy is low.=)')
elif allergy == 2:
    dust = input('Do you have dust allergy?(Write Yes or No)')
    if dust == 'Yes':
        print('Risk of dust allergy is 100% =(')
    else: 
        print('Risk of dust allergy is low.=)')
elif allergy == 3:
    pet = input('Do you have a pet?(Write Yes or No)')
    if pet == 'Yes':
        print('Risk of pet allergy is high.=(')
    else: 
        print('Risk of pet allergy is low.=)')
else:
    print('Wait for another allergy.=|')

```
